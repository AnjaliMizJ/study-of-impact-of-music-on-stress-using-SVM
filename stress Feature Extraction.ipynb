```python
#function to split the data into segments based on the time domain
def split_seq(s, size):
        newseq = []
        splitsize = 1.0/size*len(s)
        for i in range(size):
                #print(i)
               # print(splitsize)
                newseq.append(s[int(round(i*splitsize)):int(round((i+1)*splitsize))])
        return newseq
```
