```python


from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix



path=r'C:\Users\etudiant\Downloads\preprocessed data\Feature_perf_MvsF'


for i in range(len(listdir(path))):#getting all the list of channel folder 
    s=listdir(path)
    loc=(join(path,s[i]))
    d=listdir(loc)
    print(d)
    for k in range(0,5):        #accessing al the feature file file of each band for every channel
        print(len(listdir(loc)))
        print("value of ",d[1])
        loc1=join(path +'/'+ s[i],d[k])
        data=pd.read_excel(loc1)

        bnd=d[k].split('.')

        print(s[i],d[k])
        file = pd.read_excel(loc1)
    #print(file.columns.values)
        col= [x for x in file.columns if x != 'Target']
        X = file.drop(columns=['Target'])
        y = file.drop(columns=col)
    
        X_train, X_test, y_train, y_test = train_test_split(X, y) #spliiting the data into test and training
        print(X_train.shape)
        print(y_train.shape)
#Import svm model

#Create a svm Classifier
        clf = svm.SVC(kernel='poly') # Linear Kernel

#Train the model using the training sets
        clf.fit(X_train, y_train)

#Predict the response for test dataset
        y_pred = clf.predict(X_test)
    
    


    #accuracy=metrics.accuracy_score(y_test, y_pred)
    #sensitivity=metrics.recall_score(y_test, y_pred)
   
        cm1 = confusion_matrix(y_test,y_pred)
        print('Confusion Matrix : \n', cm1)

        total1=sum(sum(cm1))
#####from confusion matrix calculate accuracy
        accuracy1=(cm1[0,0]+cm1[1,1])/total1
        print ('Accuracy : ', accuracy1)

        sensitivity1 = cm1[0,0]/(cm1[0,0]+cm1[0,1])
        print('Sensitivity : ', sensitivity1 )

        specificity1 = cm1[1,1]/(cm1[1,0]+cm1[1,1])
        print('Specificity : ', specificity1)
    
    
    
        w=openpyxl.load_workbook(r"C:\Users\etudiant\Downloads\preprocessed data\svm_Feature_Perf_MvsF.xlsx")
    #writing all the outputs to the excel file 
        sheet=w.get_sheet_by_name('Feuil1')
        r=sheet.max_row
        sheet.cell(row = 1, column = 1).value="Channel"
        sheet.cell(row = 1, column = 2).value="Band"
        sheet.cell(row = 1, column = 3).value="Accuracy"
        sheet.cell(row = 1, column = 4).value="Sensitivity"
        sheet.cell(row = 1, column = 5).value="Specivity"
    
        sheet.cell(row=(r+1),column=1).value=s[i]
        sheet.cell(row=(r+1),column=2).value=bnd[0]
    
        sheet.cell(row = (r+1), column = 3).value=accuracy1
        sheet.cell(row = (r+1), column = 4).value=sensitivity1
        sheet.cell(row = (r+1), column = 5).value=specificity1

        w.save(r"C:\Users\etudiant\Downloads\preprocessed data\svm_Feature_Perf_MvsF.xlsx")
   
       
```
